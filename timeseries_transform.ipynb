```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data 경로 설정 
root = os.path.join(os.getcwd(), "drive", "MyDrive", "Colab Notebooks", "data")
```

```python
f_name = list()
for f in os.listdir(root):
  if os.path.isfile(os.path.join(root,f)):
    f_name.append(os.path.join(root,f))
for i in range(len(f_name)):
  print(i,f_name[i])
```

```python
timeseries = pd.read_csv(f_name[23])
timeseries = timeseries.drop(columns=["Unnamed: 0"])
timeseries["time"] = pd.to_datetime(timeseries["time"])
timeseries
```

```python
candidates = list(timeseries["zone"].unique())
nums = list(range(len(candidates)))
table = {}
for c, n in zip(candidates, nums):
  table[c] = n
table
```

```python
timeseries.replace({"zone": table}, inplace=True)
timeseries
```

```python
timeseries.to_csv("/content/drive/MyDrive/Colab Notebooks/data/timeseries_transform.csv", index=False)
```

```python
# unique_id : zone_name dictionary 정보

table_rvs = {v:k for k,v in table.items()}

li,li2 = [],[]
for i in range(len(table_rvs)):
  li.append(i)
  li2.append(table_rvs[i])

print(li[:5])
print(li2[:5])

col1 = li
col2 = li2
zone_information = pd.DataFrame(
    (zip(col1,col2)), 
    columns = ['unique_id','zone_name'])
zone_information

zone_information.to_csv("/content/drive/MyDrive/Colab Notebooks/data/timeseries_zone_dictionary2.csv", index=False, encoding="utf-8-sig")
```
